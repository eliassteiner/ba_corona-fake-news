```python
df_FAN = pd.read_csv("https://chruezfalsch.ch/df_FANG-COVID_cleaned.csv")
del df_FAN['Unnamed: 0'] 
df_BA = pd.read_csv("https://chruezfalsch.ch/df_ba_cleaned.csv")
del df_BA['Unnamed: 0'] 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

from yellowbrick.text import TSNEVisualizer
from yellowbrick.datasets import load_hobbies

# Load the data and create document vectors
corpus = df_BA
cv_bla= TfidfVectorizer(min_df=0.0,max_df=1.0,use_idf=True,ngram_range=(1,3))

X = cv_bla.fit_transform(corpus.text)
y = corpus.fake

# Create the visualizer and draw the vectors
tsne = TSNEVisualizer()
tsne.fit(X, y)
tsne.show()
```
